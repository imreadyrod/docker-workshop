```python
import pandas as pd
from sqlalchemy import create_engine
from tqdm.auto import tqdm
```

```python
prefix = r'https://github.com/DataTalksClub/nyc-tlc-data/releases/download/misc/'
url = f'{prefix}/taxi_zone_lookup.csv'

engine = create_engine('postgresql+psycopg://root:root@localhost:5432/ny_taxi')

chunksize = 100000

df_iter = pd.read_csv(
    url,
    iterator=True,
    chunksize=chunksize
)
target_table = 'zones'
first = True
```

```python
for df_chunk in tqdm(df_iter):

    if first:
        # Create table schema (no data)
        df_chunk.head(0).to_sql(
            name=target_table,
            con=engine,
            if_exists="replace"
        )
 
        first = False
        print("Table created")

    # Insert chunk
    df_chunk.to_sql(
        name=target_table,
        con=engine,
        if_exists="append"
    )

    print("Inserted:", len(df_chunk))
```
